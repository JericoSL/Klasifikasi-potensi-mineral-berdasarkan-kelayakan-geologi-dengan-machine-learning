```python
#Check data(features) on each column, Coordinates is drop since it won't be used
row_data = df_all_features.iloc[:,:21]

row_data.info()
```

```python
row_data.columns = df_fields['alias']

row_data.columns
```

```python
#drop attribute yang tidak dipakai (hanya sebagai catatan)
columns_to_drop = ['objectid', 'Catatan','Acuan','Metadata']
new_data = row_data.drop(columns_to_drop, axis = 'columns')
new_data
```

```python
#encoder for categorical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
datale = new_data

datale.loc[:,['Nama Komoditi Bukan Logam dan Batuan']] = le.fit_transform(new_data.loc[:,['Nama Komoditi Bukan Logam dan Batuan']])
datale.loc[:,['Kode Unsur Komoditi Bukan Logam dan Batuan']] = le.fit_transform(new_data.loc[:,['Kode Unsur Komoditi Bukan Logam dan Batuan']])
datale.loc[:,['Kelompok Komoditi Mineral Bukan Logam dan Batuan']] = le.fit_transform(new_data.loc[:,['Kelompok Komoditi Mineral Bukan Logam dan Batuan']])
datale.loc[:,['Nama Objek']] = le.fit_transform(new_data.loc[:,['Nama Objek']])
datale.loc[:,['Keterangan Nama Lokasi Keterdapatan Mineral Bukan Logam dan Batuan']] = le.fit_transform(new_data.loc[:,['Keterangan Nama Lokasi Keterdapatan Mineral Bukan Logam dan Batuan']])
datale.loc[:,['Status Penyelidikan Mineral Bukan Logam']] = le.fit_transform(new_data.loc[:,['Status Penyelidikan Mineral Bukan Logam']])
datale.loc[:,['Kabupaten']] = le.fit_transform(new_data.loc[:,['Kabupaten']])
datale.loc[:,['Provinsi']] = le.fit_transform(new_data.loc[:,['Provinsi']])
datale.loc[:,['Kecamatan']] = le.fit_transform(new_data.loc[:,['Kecamatan']])
datale.loc[:,['Pulau']] = le.fit_transform(new_data.loc[:,['Pulau']])
```

```python
#predictable based on highest correlation attribute (label encoded)
X = datale.loc[:,['Nama Objek','Kode Unsur Komoditi Bukan Logam dan Batuan' ]]
X
```

```python
#Label Y nama komoditi
y = row_data.loc[:,['Nama Komoditi Bukan Logam dan Batuan']]
y
```

```python
#train and test dataset
from sklearn.model_selection import train_test_split
from sklearn import datasets
# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#split datanya 70% untuk train dan 30% untus test
X_train
```

```python
#label Y nama komoditi
y_train
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
```

```python
#regresi logistik
logreg = LogisticRegression(solver='liblinear', random_state=42)
logreg.fit(X_train, y_train)
```

```python
y_pred_logreg = logreg.predict(X_test)

y_pred_logreg
```

```python
from sklearn.metrics import accuracy_score
#Model Accuracy of 0.6  is it acceptable? NO
print('Model accuracy score: {0:0.4}'. format(accuracy_score(y_test, y_pred_logreg)))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_logreg))
```

```python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred_gauss = gnb.predict(X_test)

y_pred_gauss
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gauss)))
```

```python
print(classification_report(y_test, y_pred_gauss))
```

```python
# K-nearest neighborh
# import KNeighbors ClaSSifier from sklearn
from sklearn.neighbors import KNeighborsClassifier


# instantiate the model
knn = KNeighborsClassifier(n_neighbors=1)


# fit the model to the training set
knn.fit(X_train, y_train)
```

```python
y_pred_knn1 = knn.predict(X_test)

y_pred_knn1
```

```python
print('Model accuracy score with k=1: {0:0.4f}'. format(accuracy_score(y_test, y_pred_knn1)))
```

```python
#Check other N_neighbors Value
knn_2 = KNeighborsClassifier(n_neighbors=2)

knn_2.fit(X_train, y_train)

y_pred_2 = knn_2.predict(X_test)

print('Model accuracy score with k=2 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_2)))
```

```python
#Check other N_neighborh Value
knn_3 = KNeighborsClassifier(n_neighbors=3)

knn_3.fit(X_train, y_train)

y_pred_3 = knn_3.predict(X_test)

print('Model accuracy score with k=3 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_3)))
```

```python
#with higher K-value the accuracy seems to be dropping
#K-value of 1 seems to give the highest accuracy, as such the report will use the K-value of 1
print(classification_report(y_test, y_pred_knn1))
```
